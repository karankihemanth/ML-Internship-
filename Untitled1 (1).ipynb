```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
```

```python
df =pd.read_csv("mcdonalds.csv")
```

```python
df.head()
```

```python
encoder = OneHotEncoder(sparse=False)
encoded_cols = encoder.fit_transform(df[['yummy', 'convenient', 'spicy', 'Gender', 'VisitFrequency']])
encoded_df = pd.DataFrame(encoded_cols, columns=encoder.get_feature_names_out())
```

```python
numeric_cols = df[['Like', 'Age']]
full_data = pd.concat([encoded_df, numeric_cols.reset_index(drop=True)], axis=1)
```

```python
plt.figure(figsize=(8, 8))
df['Gender'].value_counts().plot.pie(autopct='%1.1f%%', colors=['skyblue', 'lightgreen'], labels=['Female', 'Male'])
plt.title('Gender Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Age', y='Like', hue='Gender', data=df)
plt.title('Age vs Like Score by Gender')
plt.show()
```

```python
categorical_cols = ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap', 'tasty',
                    'expensive', 'healthy', 'disgusting', 'VisitFrequency', 'Gender']
```

```python
df.head()
```

```python
encoder = OneHotEncoder(sparse=False)
encoded_features = encoder.fit_transform(df[categorical_cols])
```

```python
def clean_like(value):
    if isinstance(value, str):  
        if 'love' in value:
            return int(value.split('!')[1])  
        elif 'hate' in value:
            return int(value.split('!')[1])
    return value 

df['Like'] = df['Like'].apply(clean_like)
print(df['Like'])


encoder = OneHotEncoder(sparse=False)
encoded_features = encoder.fit_transform(df[categorical_cols])
numerical_cols = df[['Like', 'Age']]
feature_matrix = np.hstack((encoded_features, numerical_cols))
scaler = StandardScaler()
scaled_features = scaler.fit_transform(feature_matrix)
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
true_labels = df['Gender'].map({'Female': 0, 'Male': 1}).values
predicted_labels = df['Cluster']
cm = confusion_matrix(true_labels, predicted_labels)
ConfusionMatrixDisplay(confusion_matrix=cm).plot(cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
gender_sales = df.groupby('Gender').sum()
print("Gender-wise sales:")
print(gender_sales)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
gender_sales.plot(kind='bar', color=['skyblue', 'lightgreen'])
plt.title('Gender-wise Sales')
plt.xlabel('Gender')
plt.ylabel('Total Sales (Likes)')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
